```python
def LN_data2():
    #匯入
    import pandas as pd
    import numpy as np
    #讀取
    LN_apply = pd.read_csv("C:/Users/user/Desktop/competition/TBN_LN_APPLY.csv")
    #設置月份資料以及答案
    LN_apply["month"] = "0"
    LN_apply["LN_IND"] = 1
    months = ["M1","M2","M3","M4"]
    date = [9447,9447+31,9447+61,9447+91,9568]
    LN_apply["month"] = pd.cut(LN_apply['TXN_DT'],date,right = False,labels=months)
    #切分資料
    LN_M1 = LN_apply[LN_apply["month"] == "M1"]
    LN_M2 = LN_apply[LN_apply["month"] == "M2"]
    LN_M3 = LN_apply[LN_apply["month"] == "M3"]
    LN_M4 = LN_apply[LN_apply["month"] == "M4"]
    #根據月份 創造各種LN_USE列 
    LN_pivot_M1 = pd.pivot_table(LN_M1.drop(["month","TXN_DT","LN_IND"],axis = 1).groupby(["CUST_NO","LN_USE"]).sum(),index = 'CUST_NO',columns = "LN_USE").fillna(0)
    LN_pivot_M2 = pd.pivot_table(LN_M2.drop(["month","TXN_DT","LN_IND"],axis = 1).groupby(["CUST_NO","LN_USE"]).sum(),index = 'CUST_NO',columns = "LN_USE").fillna(0)
    LN_pivot_M3 = pd.pivot_table(LN_M3.drop(["month","TXN_DT","LN_IND"],axis = 1).groupby(["CUST_NO","LN_USE"]).sum(),index = 'CUST_NO',columns = "LN_USE").fillna(0)
    LN_pivot_M4 = pd.pivot_table(LN_M4.drop(["month","TXN_DT","LN_IND"],axis = 1).groupby(["CUST_NO","LN_USE"]).sum(),index = 'CUST_NO',columns = "LN_USE").fillna(0)
    #存取各月份LN_USE資料
    USE_M1 = LN_pivot_M1.columns.droplevel()
    USE_M2 = LN_pivot_M2.columns.droplevel()
    USE_M3 = LN_pivot_M3.columns.droplevel()
    USE_M4 = LN_pivot_M4.columns.droplevel()
    USE_temp = USE_M1.append(USE_M3).drop_duplicates() 
    #創建空DF
    column_M1 = pd.DataFrame()
    column_M2 = pd.DataFrame()
    column_M3 = pd.DataFrame()
    column_M4 = pd.DataFrame()
    columns_all_M1 = pd.DataFrame()
    columns_all_M2 = pd.DataFrame()

    #命名欄位
    for USE in USE_M1:
        column_M1['LN_AMT_M1_'+USE] = 0   
    for USE in USE_M2:
        column_M2['LN_AMT_M2_'+USE] = 0
    #為了機器學習擁有相同變數 在此設置相同名稱
    for USE in USE_M3:
        column_M3['LN_AMT_M1_'+USE] = 0
    for USE in USE_M4:
        column_M4['LN_AMT_M2_'+USE] = 0
    for USE in USE_temp:
        columns_all_M1["LN_AMT_M1_"+USE] = None
        columns_all_M2["LN_AMT_M2_"+USE] = None
    #合併欄位
    LN_pivot_M1.columns = column_M1.columns
    LN_pivot_M2.columns = column_M2.columns
    LN_pivot_M3.columns = column_M3.columns
    LN_pivot_M4.columns = column_M4.columns

    columns_all_M1.index = columns_all_M1.index.set_names(["CUST_NO"])
    columns_all_M2.index = columns_all_M2.index.set_names(["CUST_NO"])

    LN_M1 = columns_all_M1.append(LN_pivot_M1).fillna(0)
    LN_M2 = columns_all_M2.append(LN_pivot_M2).fillna(0)
    LN_M3 = columns_all_M1.append(LN_pivot_M3).fillna(0)
    LN_M4 = columns_all_M2.append(LN_pivot_M4).fillna(0)

    LN_M1["LN_IND_M1"] = 1
    LN_M2["LN_IND_M2"] = 1
    LN_M3["LN_IND_M1"] = 1
    LN_M4["LN_IND_M2"] = 1

    LN_M1 = LN_M1.reset_index()
    LN_M2 = LN_M2.reset_index()
    LN_M3 = LN_M3.reset_index()
    LN_M4 = LN_M4.reset_index()
    return LN_M1,LN_M2,LN_M3,LN_M4
```
